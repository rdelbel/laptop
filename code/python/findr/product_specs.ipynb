```python
def get_page_of_urls(url):
    page=parse_url(url)
    links=page.find_class("thumb")
    return [link.get('href') for link in links]
```

```python
def get_next_page(url):
    url=parse_url(url).find_class("next_page")[0].get('href')
    if(type(url)==str):
        return 'http://www.theverge.com'+url
    else:
        return -1

    
```

```python
get_next_page('http://www.theverge.com/products/categories/laptops/6')
```

```python
def get_all_urls(url):
    urls=[url]
    url=get_next_page(url)
    while(url!=-1):
        urls.append(url)
        url=get_next_page(url)
    return urls
```

```python
urls=get_all_urls('http://www.theverge.com/products/categories/laptops/6')
```
